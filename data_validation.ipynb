```python
import pandas as pd
import numpy as np
```

```python
# loading datasets
campaigns = pd.read_csv('datasets/campaigns.csv')
customers = pd.read_csv('datasets/customers.csv')
events = pd.read_csv('datasets/events.csv')
products = pd.read_csv('datasets/products.csv')
transactions = pd.read_csv('datasets/transactions.csv')
```

```python
# cleaning datasets
datasets = {
    "campaigns": campaigns,
    "customers": customers,
    "events": events,
    "products": products,
    "transactions": transactions
}

for name, df in datasets.items():
    print(f"The shape of {name} dataset is {df.shape}")
    print(f"The missing values in {name} dataset is {df.isna().sum()}")
    print(f"The duplicate values in {name} is {df.duplicated().sum()}")
    print()

print("The Missing values are in EVENTS and TRANSACTIONS dataset ")
```

```python
# cleaning missing values from events
print((events.isna().sum())/len(events) * 100)
```

```python
events.dropna(inplace=True)
```

```python
events.shape
```

```python
# cleaning missing values from Transactions
print((transactions.isna().sum()/len(transactions)) * 100)
```

```python
transactions.dropna(inplace=True)
```

```python
transactions.shape
```

```python
datasets = {
    "campaigns": campaigns,
    "customers": customers,
    "events": events,
    "products": products,
    "transactions": transactions
}

for name, df in datasets.items():
    print(f"The info of {name} dataset is {df.info()}")
    print()

print("The Product_id from EVENTS, TRANSACTIONS is Float")
```

```python
events['product_id'] = events['product_id'].astype('Int64')
transactions['product_id'] = transactions['product_id'].astype('Int64')
```
